```python
import numpy as np
from nnlib.layers import Dense, ReLU, Softmax
from nnlib.losses import CrossEntropy
from nnlib.model import Sequential
from nnlib.optims import SGD
```

```python
np.random.seed(42)

X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
Y = np.array([
    [1, 0],
    [0, 1],
    [0, 1],
    [1, 0]
])  # XOR problem
```

```python
# define the model
model = Sequential([
    Dense(2, 32),
    ReLU(),
    Dense(32, 16),
    ReLU(),
    Dense(16, 2),
    Softmax()
])
```

```python
# define model hyper parameters
lr = 0.001
epochs = 1000
optim = SGD(lr=lr)
loss_fn = CrossEntropy()
```

```python
# training the model
for _ in range(epochs):
    Y_pred = model.forward(X)
    loss = loss_fn.forward(Y, Y_pred)
    grad = loss_fn.backward()
    model.backward(grad)
    optim.update(model.parameters())
```

```python
print("Final predictions:", np.round(model.forward(X)))
```
